```python
import pandas as pd
import matplotlib.pyplot as plt



df = pd.read_csv('vgsales.csv' , encoding='UTF-8')



rows = (df['Publisher'] == 'Nintendo')
columns = [ 'Name' , 'Global_Sales']
result = df.loc[rows , columns].head(10)
result.set_index('Name', inplace=True)
print(result)

chart = result.plot(kind='bar',
     title='Nintendo Top 10 Sales', 
     xlabel='name',
     ylabel='globle sales',
     legend=True,
     figsize=(10,5))
plt.show()
```
